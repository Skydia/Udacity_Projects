```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from IPython.display import display
%matplotlib inline

COLOR = 'white'
mpl.rcParams['text.color'] = COLOR
mpl.rcParams['axes.labelcolor'] = COLOR
mpl.rcParams['xtick.color'] = COLOR
mpl.rcParams['ytick.color'] = COLOR
pd.options.display.max_columns = None

df = pd.read_csv("2017-fordgobike-tripdata.csv")
df.head()
```

```python
df['start_time'] = pd.to_datetime(df['start_time'], format='%Y-%m-%d %H:%M:%S')
df['end_time'] = pd.to_datetime(df['end_time'], format='%Y-%m-%d %H:%M:%S')
df.info()
```

# Ford GoBike User Behavior

# Investigation Overview
In this investigation, I would like to see the behavior of customers based on user type, station location, distance, duration, and months.

# Dataset Overview
This dataset consisted of 519700 rows and 13 columns. Data types are mixed including object, integer, float, and datetime. There are 272 Stations with 3673 bikes. On the 'user_type' column, there are two unique values which are 'Customer' and 'Subscriber'. 8.91% rows were removed due to outliers and unwanted conditional variable.

# Distribution of Rental Duration in 2017
Duration in this dataset is from 0 to 80k+ seconds. The interquartile range is between 382 and 938. X-Axis scale is changed into logarithmic scale, therefore it's seen as normal distribution.

```python
# Another plot in log scale
# Create figure
fig = plt.figure()

# specify bin size and maximum values
log_binsize = 1/100
bins = 10 ** np.arange(np.log10(df['duration_sec'].min()), np.log10(df['duration_sec'].max())+log_binsize, log_binsize)

# Specify 0.25 and 0.75 percentiles
q25, q75 = np.percentile(df['duration_sec'], [25 ,75])

# Plot histogram
plt.hist(data=df, x='duration_sec', bins=bins)
plt.axvline(x=q25, color='orange', label='25 Percentile', linestyle='--')
plt.axvline(x=q75, color='red', label='75 Percentile', linestyle='--')

# Plot texts
plt.text(q25-200, 3950, int(q25), color='black')
plt.text(q75+100, 3950, int(q75), color='black')

# Set scale
plt.xscale('log')

# Set ticks
plt.xticks([100, 300, 500, 1e3, 2e3, 5e3, 1e4, 2e4, 4e4, 8e4], 
           ['100','300','500', '1k', '2k', '5k', '10k', '20k', '40k', '80k'])

# Set labels
plt.xlabel('Duration (second)')
plt.ylabel('Amount')
plt.title('Amount of Customers Rental Duration in 2017')
leg = plt.legend()
for line, text in zip(leg.get_lines(), leg.get_texts()):
    text.set_color(line.get_color())
```

# Distribution of Monthly Rentals in 2017
The highest rentals amount is in October. Note that months available started from June until December and the amount of rental in June is very low compared to the other months.

```python
df['month'] = df['start_time'].dt.month
# set default color
default_color = sns.color_palette()[0]

# plot using seaborn countplot
ax = sns.countplot(data = df, x='month', color=default_color)

# set title and label name
ax.set_xlabel("Month (Number)")
ax.set_ylabel("Amount of Rentals")
ax.set_title("Amount of Monthly Rentals in 2017");
```

# Distribution of Rentals in each User Type
Most of the rentals are occured by the Subscriber. This is unexpected and a good news for the company.

```python
plt.figure(figsize=(8,8))

# set default color
default_color = sns.color_palette()[0]

# plot using seaborn countplot
ax = sns.countplot(data = df, x='user_type', color=default_color)

# set title and label name
ax.set_xlabel("User Type")
ax.set_ylabel("Amount of Rentals")
ax.set_title("Total Rents of User Type Category in 2017")
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.tight_layout();
```

```python
Q1 = df['duration_sec'].quantile(0.25)
Q3 = df['duration_sec'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - (1.5 * IQR)
upper = Q3 + (1.5 * IQR)
new_df = df[(df['duration_sec']>=lower) & (df['duration_sec']<=upper)].reset_index(drop=True)
new_df.head()
```

```python
def find_distance(start_lat,start_lon,end_lat,end_lon):
    from geopy.distance import distance
    
    coords_1 = (start_lat, start_lon)
    coords_2 = (end_lat, end_lon)

    return distance(coords_1, coords_2).m

new_df['distance']= np.vectorize(find_distance)(new_df['start_station_latitude'], 
                                                new_df['start_station_longitude'], 
                                                new_df['end_station_latitude'], 
                                                new_df['end_station_longitude'])
new_df.head()
```

# Duration (Second) vs Distance (Meter)
There is a relationship between Duration and Distance. Pearson correlation coeffiecient value (of 0.73) is above than 0.5. Therefore, these two variables has a high correlation on each other.

```python
above_0_dist_df = new_df.query("distance > 0")
# Scatter Plot of duration_sec and distance with changed alpha
# set default color
default_color = sns.color_palette()[0]

# plot using seaborn countplot
ax = sns.regplot(data = above_0_dist_df, 
                 x='duration_sec', 
                 y='distance', 
                 color=default_color, 
                 scatter_kws={'alpha':0.01, 'color':'red'})

# set title and label name
ax.set_xlabel("Duration (Second)")
ax.set_ylabel("Distance (Meters)")
ax.set_title("Distance(Meters) vs Duration(Second) Distrubution in 2017");
```

# Duration and Distance by User Type
Customer type tends to rent longer than Subscriber. This is shown by all kind of plots including its average. Customer type also tends to have higher distance as well based Box Plot and Bar Chart. While from the Violin Plot, it tells us that the distribution is similar between two user types, except that Subscriber Type have a larger range.  
The duration_sec distribution of Customer type seems to be more distributed too, compared to Subscriber type that have a lower median.

```python
# Create figure
f, axes = plt.subplots(1, 3, figsize=(20, 15))

# set default color
default_color = sns.color_palette()[0]

# plot using seaborn
sns.boxplot(data=above_0_dist_df, x='user_type', y='duration_sec',ax=axes[0], color=default_color)
sns.violinplot(data=above_0_dist_df, x='user_type',y='duration_sec', ax=axes[1], color=default_color)
sns.barplot(data=above_0_dist_df, x='user_type', y='duration_sec',ax=axes[2], color=default_color)

axes[0].set_title("User Type vs Duration (Second) Boxplot in 2017")
axes[1].set_title("User Type vs Duration (Second) Violinplot in 2017")
axes[2].set_title("User Type vs Duration (Second) Barchart in 2017")
axes[0].set_xlabel("User Type")
axes[1].set_xlabel("User Type")
axes[2].set_xlabel("User Type")
axes[0].set_ylabel("Duration (second)")
axes[1].set_ylabel("Duration (second)")
axes[2].set_ylabel("Average Duration (second)")
axes[2].set_ylim((0, 2000));
```

```python
# Create figure
f, axes = plt.subplots(1, 3, figsize=(20, 15))

# set default color
default_color = sns.color_palette()[0]

# plot using seaborn
sns.boxplot(data=above_0_dist_df, x='user_type', y='distance',ax=axes[0], color=default_color)
sns.violinplot(data=above_0_dist_df, x='user_type',y='distance', ax=axes[1], color=default_color)
sns.barplot(data=above_0_dist_df, x='user_type', y='distance',ax=axes[2], color=default_color)

axes[0].set_title("User Type vs Duration (Second) Boxplot in 2017")
axes[1].set_title("User Type vs Duration (Second) Violinplot in 2017")
axes[2].set_title("User Type vs Duration (Second) Barchart in 2017")
axes[0].set_xlabel("User Type")
axes[1].set_xlabel("User Type")
axes[2].set_xlabel("User Type")
axes[0].set_ylabel("Distance")
axes[1].set_ylabel("Distance")
axes[2].set_ylabel("Average Distance (meter)")
axes[2].set_ylim((0, 2000));
```

# Duration by Months for Each User Type
It seems there aren't significant difference between months on each user type. We still found the same thing that Customer tends to rent longer than Subscriber.
The confidence interval of this pointplot in June seems to be much larger than the other months. It's caused by the number of rows in June seems to be very low compared to the other months.

```python
plt.figure(figsize=(10,10))
sns.pointplot(x='month', y='duration_sec', hue='user_type', data=above_0_dist_df, linestyles = '')
plt.xlabel('Month')
plt.ylabel('Duration(Second)')
leg = plt.legend()
for text in leg.get_texts():
    text.set_color("black")
plt.title('Monthly Average Duration(Second) by User Type in 2017');
```

# Distance by Month for Each User Type
This tells us differently compared to the previous plot. It tends to get higher every month based on the above plot. This plot shows that people doesn't ride far in the summer season.  
In the Autumn season, people would likely to ride further compared to in the summer season. But this doesn't prove on the duration though. In average, people tends to ride in about 15 minutes no matter in which months.

```python
plt.figure(figsize=(10,10))
sns.pointplot(x='month', y='distance', hue='user_type', data=above_0_dist_df, linestyles = '', dodge=True)
plt.xlabel('Month')
plt.ylabel('Average Distance(Meters)')
leg = plt.legend()
for text in leg.get_texts():
    text.set_color("black")
plt.title('Monthly Average Distance(Meters) by User Type in 2017');
```
